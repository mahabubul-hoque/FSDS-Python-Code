### Assignment 1.1: Temperature Check 
Problem: 
Write a program that takes the current temperature as input and prints:  
● It's too cold!" if below 10°C  
● It's cool outside" if between 10°C and 25°C  
● It's hot outside!" if above 25°C  

```python
temp = float(input("Enter current tempature: "))

if temp < 10:
    print("It is too cold!")

elif temp >= 10 and temp <=25:
    print("It's cool outside!")
    
else:
    print("it's too hot!") 

```

### Assignment 1.2: ATM Withdrawal System  
Problem: Ask the user for their account balance and the amount they want to withdraw.  
● If the withdrawal amount is greater than the balance → print "Insufficient Balance"   
● If the withdrawal amount is less than or equal to balance → print "Transaction Successful" and show the remaining balance.  

```python
balance = float(input("What is your current balance"))
withdraw = float(input("How much you want to withdraw"))

if withdraw > balance:
    print("Insufficient Balance")
else:
    print("Trasaction Succeful, Your current balance is ", balance - withdraw)
```

### Assignment 1.3: Student Grade Calculator  
Problem: Input marks (0–100) and print grade based on:  
● 90–100 → A+  
● 80–89 → A  
● 70–79 → B  
● 60–69 → C  
● Below 60 → Fail  

```python
mark = int(input("Input marks (0 – 100): "))
if mark >= 90 and mark <=100:
    print("A+")
elif mark >= 80 and mark <=89:
    print("A")
elif mark >= 70 and mark <=79:
    print("B")
elif mark >= 60 and mark <=69:
    print("C")
elif mark < 60:
    print("Fail")
else:
    print("Something wrong")

```

### Assignment 1. 4: Bus Ticket Discount  
Problem: Ask the user for age and print ticket price:  
● Below 5 → Free  
● 5–18 → 50% Discount

● Others → Full Price  

```python
age = int(input("Please enter your age in year: "))

if age <5:
    print("You are entitled FREE Ticket")

elif age >= 5 and age <=18:
    print("You are allowed 50% discount' on Ticket")

elif age > 18 and age <= 60:
    print("You are allowed 30% discount")

else:
    print("Full Price")
```

### Assignment 1.8: Leap Year Checker  
Problem: Take a year as input and check:   
● If it is divisible by 400 → Leap Year  
● Else if divisible by 4 but not by 100 → Leap Year  
● Otherwise → Not a Leap Year  

```python
year = int(input("Enter a year: "))

if year % 400 ==0:
    print("Leap Year")
elif year % 4 ==0 and year % 100 !=0:
    print("Leap Year")
else:
    print("Not a Leap year") 
```

### Assignment 10: Electricity Bill Calculator
Problem: Input number of units consumed and calculate bill:  
● Up to 100 units → 5 BDT per unit  
● 101–200 units → 7 BDT per unit  
● Above 200 units → 10 BDT per unit   
Print total payable amount.  

```python
unit = int(input("Enter unit of electricity consumed: "))

if unit <= 100:
    print("Your total payment amount of electricity bill is BDT", unit*5)
elif unit >= 101 and unit <= 200:
    print("Your total payable amount of electricity bill is BDT", 100*5+(unit-100)*7)
else:
    print("Your total payable amount of electricity bill is BDT", 100*5+100*7+(unit-100-100)*10 )
```
