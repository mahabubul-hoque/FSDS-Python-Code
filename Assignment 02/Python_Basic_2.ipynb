### 2.2. Count Word Frequency in Customer Reviews 
### Real-world use: Analyzing reviews or comments in text analytics.  
Task: Given a review text, count how many times each word appears (ignore case). 
Example:  
Input: "This product is good and this service is excellent"  
Output: {'this': 2, 'product': 1, 'is': 2, 'good': 1, 'and': 1, 'service': 1, 'excellent': 1}  

```python
# Input text
text = input(f"Enter a review: ")

textl = text.lower() # Convert all text in lower case
words = textl.split() # Split into words
word_count = {} # Create an empty dictionary to store word counts

for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

print(text)
print(word_count)
```

### 2.3. Password Strength Checker 
### Real-world use: Used in user authentication systems.  
Task: Check if a password is strong — must have at least one uppercase, one lowercase, one digit, and one special character.  
Example:  
Input: "Hello@123"  
Output: Strong Password  

```python
password = input("Enter password: ")

upper = lower = digit = special = 0
special_chars = "!@#$%^&*(),.?\":{}|<>"

for char in password:
    if char.isupper():
        upper = 1
    if char.islower():
        lower = 1
    if char.isdigit():
        digit = 1
    if char in special_chars:
        special = 1

if upper and lower and digit and special:
    print("Strong Password")
else:
    print("Weak Password")
```

### 2.4. Username Generator
### Real-world use: Automatically create usernames for users.
Task: Take user’s full name and generate a short username.   
Rules: first 3 letters of first name + last 3 letters of last name + random number (use loops only).   
Example:   
Input: "Mamun Malitha"   
Output: mamitha57   

```python
import random

number = random.randint(1, 100)
full_name = input("Enter your full name: ")

first_name = full_name[:3].lower()
last_name = full_name[-3:].lower()

userid = first_name + last_name + str(number)

trimmedid = userid.replace(" ", "") # remove space if any.

print("Your Name: :",full_name)
print("Generated User Name:",trimmedid)
```

### 2.5. SMS Word Counter 
### Real-world use: Used in messaging apps or social media post limit check. 
Task: Count how many characters and words are in a message, and tell if it exceeds 160 characters (SMS limit).  
Example:  
Input: "This is a demo message."  
Output: Words = 5, Characters = 23, Status = Within Limit  

```python
message = input("Please enter a message: ")
char_count = len(message)
words = message.split()
word_count = len(words)

if char_count >160:
    limit = "Limit Exceeded"
else:
    limit = "Within Limit"

print("Your message:",message)
print(f"Words = {word_count}, Characters = {char_count}, Status = {limit}")
```

### 2.6. Remove Duplicates from Contact List  
### Real-world use: Data cleaning in CRM or user databases. 
Task: Given a list of names (comma separated), remove duplicates.  
Example:  
Input: "Rahim, Karim, Rahim, Sultana"  
Output: "Rahim, Karim, Sultana"  

```python
input_str = "Rahim, Karim, Rahim, Sultana"

names = input_str.split(", ") # Split the string into a list

unique_names = [] # Check and add unique names in the list
for name in names:
    if name not in unique_names:
        unique_names.append(name)

output_str = ", ".join(unique_names) # Join the list by "," to a string
print(output_str)
```

### 2.7. Detect Suspicious Words in Text (Moderation)
### Real-world use: Used in chat moderation or email filtering.
Task: If any “banned” words like ['spam', 'scam', 'fake'] are found in a text, print “Warning”.  
Example:  
Input: "This is a scam offer"  
Output: Warning: Suspicious content detected  

```python
body = input("Enter the chat or email to check warning or harmful text: ")
banned = ['spap', 'scam', 'fake']
ask = input("Do you want to add any banned word?: Yes/No").lower()

if ask == "yes":    # update banned related words if required.
  words = input("Enter you word please:")
  banned.append(words)

body_words = body.lower().split() # Split the chat or email text to check

for i in body_words:
   for j in banned:
        if i == j:
            print(f"Warning: Suspicious content detected: {j}")
```

### 2.9. Count Hashtags and Mentions (Social Media Analytics) 
### Real-world use: Used in analyzing social media posts. 
Task: Count how many hashtags (#) and mentions (@) appear in a post.  
Example:  
Input: "Check out #AIWorkshop and follow @inceptionbd"  
Output: Hashtags = 1, Mentions = 1  

```python
post = input("Enter the complete post wtih Hashtags and Mentions: ")
hashtag = 0
mention = 0

words = post.split()

for tag in words:
    if tag.startswith("#"):
        hashtag += 1

for tag in words:
    if tag.startswith("@"):
        mention += 1

print(f"Hashtags = {hashtag}, Mentions = {mention}")

```

### 2.10. Word Reversal for Data Encryption 
### Real-world use: Simple text obfuscation or playful encoding. 
Task: Reverse each word in a sentence while keeping word order the same.  
Example:   
Input: "AI is fun"   
Output: "IA si nuf"   

```python
sentence = input("Enter a sentence: ")
words = sentence.split()

for word in words:
    reverse = word[::-1]
    print(reverse, end=" ")
```
